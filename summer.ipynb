```python
from sklearn.linear_model import LinearRegression
```

```python
import pandas
```

```python
x_train = pandas.read_csv('training_set_features.csv')
x_test = pandas.read_csv('test_set_features.csv')
y_train = pandas.read_csv('training_set_labels.csv')
y_test = pandas.read_csv('submission_format.csv')
```

```python
x_train =  x_train.drop('respondent_id', axis = 1)
```

```python
x_test =  x_test.drop('respondent_id', axis = 1)
y_train =  y_train.drop('respondent_id', axis = 1)
Y_test =  y_test.drop('respondent_id', axis = 1)
```

```python
x_train = x_train.fillna(0)
x_test = x_test.fillna(0)
```

```python
x_train = x_train.drop(['household_adults', 'household_children', 'employment_industry', 'employment_occupation'], axis = 1)
x_train.head()
```

```python
def convert(col):
    dic = {}
    j = 0
    for i in x_train.loc[:, col]:
        if i not in dic:
            dic[i] = len(dic)
        x_train.loc[j, col] = dic[i]
        j += 1
    j = 0
    for i in x_test.loc[:, col]:
        x_test.loc[j, col] = dic[i]
        j += 1
```

```python
convert('sex')
convert('race')
convert('rent_or_own')
convert('age_group')
convert('employment_status')
convert('hhs_geo_region')
convert('census_msa')
convert('education')
convert('income_poverty')
convert('marital_status')
x_train.head()
```

```python
x_test = x_test.drop(['household_adults', 'household_children', 'employment_industry', 'employment_occupation'], axis = 1)
x_test.head()
```

```python
model = LinearRegression()
model.fit(x_train, y_train)
predictions = model.predict(x_test)
```

```python
import numpy as np
predictions = np.round_(predictions, 2)
predictions = abs(predictions)
print(predictions)
```

```python
predictions = list(predictions)
index = 26707
for i in range(len(predictions)):
    predictions[i] = list(predictions[i])
    predictions[i].insert(0, index)
    index += 1
```

```python
predictions = pandas.DataFrame(predictions)
predictions
```

```python
predictions.columns = ['respondent_id', 'h1n1_vaccine', 'seasonal_vaccine']
```
